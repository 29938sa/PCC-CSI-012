# What is Object Oriented Programming (OOP)?


## The Need for OOP

- As you write more code, you'll notice that some code looks better than others
    - Sometime I find myself feeling "weird" about some code that I just wrote
- Furthermore, when you start working with larger code bases, you'll notice it gets harder to understand
- A different paradigm, or thinking is necessary to handle these types of situations

## OOP is a paradigm
- OOP is a programming paradigm where you use objects to represent specific elements within a program
- OOP is a concept to break up large code bases into more bite sized chunks or zones for developers and users
- OOP allows for hiding complexity, we don't need to know how a library work, just how to use it
- OOP allows to focus on parts of the program that we care about and then ignore the rest
    - As a design philosophy, we should try to break up the code into the smallest components as possible

## OOP isn't new
- OOP's origins go all the way back to the 1950's and 1960's
- Simula, a programming language released in the early 1970's embraced these object-oriented designs and was used researchers for modeling.
- Today, There are lot's of languages that are built around OOP (Python, Java, C++, or C# for example) or have implemented aspects of OOP (PHP, Perl, or Matlab for example)

## OOP is broken up into 4 major concepts

1. **Encapsulation:** The process of encapsulating, or organizing code so that only the relevant parts are available to the user.
1. **Abstraction:** The process of abstracting, or templating code to show the basic requirements a specific set of code should accomplish.
1. **Inheritance:**: The process of inheriting, or reusing code with little to no changes in between objects.
1. **Polymorphism:** The concept that one thing, a class, can be used to describe a multiple similiar things even though the details may be different for each.

## Pokemon, a conceptual example



Let's take a Pokemon card and examine it for a minute. What are the attributes it has? What are it's actions it can take? What if we heal a pokemon? What happens to it if it's attacked by an attack type it's weak to?

- Specific pokemon are all similar, you might say their in the same class of monsters
    - Each pokemon have similar features, or attributes and actions
- We can think of a specific pokemon as an object
    - Each pokemon have specific values for it's attributes and actions
- Each instance of Pokemon reacts differently depending on it's attributes
- Some Pokemon are related to each other via evolution or breeding
