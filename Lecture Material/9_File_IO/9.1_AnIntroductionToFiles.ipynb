# 9.1 An Introduction to Files

## Outline:
- What is a file?
    - Everything on a Computer is 1's and 0's, which are collected together as bytes, which are translations of human readable (text files, csv, xml, etc) and binary (executable, image, video, music, etc) things.  These things are typically used by programs to present data to the user.
        - Example .docx file
    - Files are stored on Secondary Memory locations (HDD, Flash Drives, etc,.) which are persistent meaning after a program is closed or a computer is restarted the file persists or isn't lost.
    - At its core, a file is: Header (optional), Contents, EOF
    - extensions are used to indicate what a file is intended to do (but not garaunteed).  For example, the PNG file extension is actually a rather well defined system:
        - https://en.wikipedia.org/wiki/Portable_Network_Graphics
        - Examine a PNG file in a hexdump
- File Paths
    - "paths" are very much like "trails", they're specific checkpoints that you traverse to get to a destination
    - paths consist of:
        - Folder Path
        - File Name
        - Extension
    - Absolute Paths
    - Relative Paths
- Line Endings
    - A fun history lesson
- Character Encoding
    - ASCII and UNICODE
